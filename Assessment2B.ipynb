###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


```python
df_exced_expect = df[df['CodeKata Score'] > 15000]
```

###### if  10000<codekata score<15000   (Reached_expectations.csv)



```python
df_reach_expect = df[(df['CodeKata Score'] > 10000) & (df['CodeKata Score'] < 15000)]
df_reach_expect.sort_values(by =['CodeKata Score'], ascending = False)
```

```python
df_reach_expect.to_csv('Reached_expectations.csv', index = False)
```

###### Department wise codekata performence (pie chart)

```python
df.Department.unique()
```

```python
df.groupby(['Department'])['CodeKata Score'].mean()
```

```python
list(df.groupby(['Department'])['CodeKata Score'].mean())
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df.Department.unique()
sizes = list(df.groupby(['Department'])['CodeKata Score'].mean())
explode = (0, 0.1, 0)  # only "explode" the 2nd slice 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
