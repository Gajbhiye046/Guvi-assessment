#### Registration and Login system using Python, file handling

1. Stage -- 1 : Registration

    a. When the user chooses to Register
        email/username should have "@" and followed by "."
        eg:- sherlock@gmail.com
             nothing@yahoo.in

    b. it should not be like this 
        eg:- @gmail.com

    c. there should not be any "." immediate next to "@"
        eg:- my@.in

    d. it should not start with special characters and numbers
        eg:- 123#@gmail.com

    e. password (5 < password length > 16)
       * Must have minimum one special character,
       * one digit,
       * one uppercase, 
       * one lowercase character 

2. Stage 2 
    Once the username and password are validated, store that data in a file


3. Stage 3 : Login
    1. When the user chooses to Login, check whether his/her credentials exist in the file or not based on the user input. 
    2. If it doesn’t exist then ask them to go for Registration or If they have chosen forget password you should be able to retrieve their original password based on their username provided in the user input or else you can ask them to provide a new password(only if their username matches with the data exists in the file).
    3. If nothing matches in your file you should ask them to Registration(Since they don't have an account)
Note:- This task has to be accomplished using Python alone. No web.


```python
import re
def check_key(dictionary, key):
     
    if key in dictionary.keys():
        value = dictionary[key]
        return "Present"
    else:
        return "Not Present"
def find_key(dictionary, key):
     
    if key in dictionary.keys():
        value = dictionary[key]
        return value
    else:
        return None
    
def checkmatch_user(string, pattern = '\D+@\D+.\D+'):
    result = re.match (pattern, string)
    return result

def checkmatch_pass(string, pattern = '[a-z0-9A-Z~`!@#$%^&*()_+=]{5,16}'):
    result = re.match (pattern, string)
    return result


def Registration(User):
    while True :
        print("Enter valid username and password with one special charcter, one digit,one lower & one upper case character")
        username = input('Enter the Username :')
        password = input('Enter the password :')
        if checkmatch_user(username) != None : #Matched the string 
            if check_key(User,username) == "Not Present" : #The username is not already present 
                if checkmatch_pass(password) : #Check whether password satisfy the condition
                    User[username] = password  # Add username and password to User dictionary
                    break

def login(User):
    username = input('Enter the Username :')
    password = input('Enter the password :')
    if check_key(User,username) == "Not Present" :
        print("Username not present . Proceed with new Registration! ")
        Registration(User)

def forgot_password(User): 
    username = input('Enter the Username :')
    req = input('Do you want to reset your password: yes/no\n')
    if req == 'no':
        pass_requested = find_key(User,username)
        return pass_requested
    else:
        password = input('Enter the new password :')
        if checkmatch_pass(password) : #The username is not already present 
            User[username] = password # Add new password to existing username.
        
```
